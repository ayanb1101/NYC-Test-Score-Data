Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend.

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell
import pandas as pd
```

```python
# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
# 80% of 800 = 640
var_df = schools[(schools["average_math"]>=640)][["school_name", "average_math"]]
best_math_schools= var_df.sort_values(by="average_math", ascending=False)
best_math_schools
```

```python
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
top_10_schools = schools[['school_name', 'total_SAT']]
top_10_schools = top_10_schools.sort_values(by='total_SAT', ascending=False).head(10)
top_10_schools
```

```python
# Calculate average SAT, standard deviation, and deviation for each school
schools["average_SAT"] = schools["total_SAT"].mean()
schools["std_SAT"] = schools["total_SAT"].std()
schools["deviation"] = (schools["total_SAT"] - schools["average_SAT"]) / schools["std_SAT"]

# Find the school with the largest standard deviation
largest_standard_deviation = schools["deviation"].idxmax()

# Initialize largest_std_dev DataFrame
largest_std_dev = pd.DataFrame()

# Get the borough with the largest standard deviation
largest_std_dev["borough"] = [schools.loc[largest_standard_deviation, "borough"]]

# Count the number of schools in that borough
largest_std_dev["num_schools"] = len(schools[schools["borough"] == largest_std_dev["borough"][0]])

# Add the average SAT and standard deviation (calculated once)
# Fix: Ensure we are calculating the mean for the correct borough
average_SAT_borough = schools[schools["borough"] == largest_std_dev["borough"][0]]["total_SAT"].mean()
largest_std_dev["average_SAT"] = round(average_SAT_borough, 2)

# Use the first value in the column as average SAT
std_SAT_borough = schools[schools["borough"] == largest_std_dev["borough"][0]]["total_SAT"].std()
largest_std_dev["std_SAT"] = round(std_SAT_borough, 2) # Use the first value in the column as standard deviation

# Display the result
largest_std_dev
```
